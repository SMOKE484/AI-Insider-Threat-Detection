```python
import pandas as pd

df = pd.read_csv('../data/insider_threat_clean_dataset.csv')
print(df.shape)
df.head()
```

```python
print(df.isnull().sum())
```

```python
print(df['is_malicious'].value_counts())
print(df['is_malicious'].value_counts(normalize=True))
```

```python
columns_to_drop = ['late_exit_flag', 'trip_day_number', 'has_medical_history', 'employee_origin_country']
df = df.drop(columns=columns_to_drop)
print("Remaining columns:", df.columns.tolist())
```

```python
from sklearn.preprocessing import LabelEncoder

categorical_cols = ['employee_department', 'employee_campus', 'employee_position']
le = LabelEncoder()

for col in categorical_cols:
    df[col] = le.fit_transform(df[col])

df.head()
```

```python
df.to_csv('../data/cleaned_dataset.csv', index=False)
print("Cleaned dataset saved!")
print("Final shape:", df.shape)
```
